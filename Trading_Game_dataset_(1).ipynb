Télécharger les données

```python
tickers = ["E5H.SI","EB5.SI","P8Z.SI","5JS.SI","AALI.JK","LSIP.JK","SIMP.JK","SMAR.JK","TBLA.JK","UNSP.JK","SSMS.JK","BWPT.JK","DBA","MOO","RJA"]

data = yf.download(tickers, start="2015-01-01")["Close"]

data.head()
```

Vérifier les données manquantes

```python
data.isna().sum()
```

```python
data = data.dropna()
data.isna().sum()
```

```python
print("Dataset shape:", data.shape)
print("Date range:", data.index.min(), "to", data.index.max())
```

```python
data.describe()
```

```python
data.to_csv("palm_oil_dataset1.csv", index=True)
```
